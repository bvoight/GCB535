**Part 3: Downloading Files From ENCODE to CoCalc**

For this module's in-class assignment, you will be working with narrowPeak bed files. For more information on this file type, see https://genome.ucsc.edu/FAQ/FAQformat.html#format12.

In the in-class assignment, you will need to find these files on ENCODE, download them to CoCalc, and unzip them.
Below is an example of how this will be done:

You will be provided with a file ID, such as **ENCFF787EAZ**.

Go to http://www.mouseencode.org/, and click on Data Search.

Search for the specific file by typing it into the search bar in the top right corner.

On the search results page, click on the bed (narrowPeak) file.

This will take you to the file summary page. Right click on the download button and copy the link address.
